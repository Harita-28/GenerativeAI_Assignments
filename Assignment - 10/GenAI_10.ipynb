```python
import torch
from diffusers import StableDiffusionPipeline

# Load the pre-trained Stable Diffusion model from Hugging Face
model_id = "runwayml/stable-diffusion-v1-5"  # You can replace this with another model
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")  # Use GPU for faster generation (if available)

# Input prompt
prompt = "A mountain surrounded by rivers"

# Generate image
image = pipe(prompt).images[0]

# Save and display the image
image.save("generated_image.png")
image.show()
```
